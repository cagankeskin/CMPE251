```python
import warnings
warnings.filterwarnings('ignore')
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
G = nx.DiGraph()

n0 = 4
layer0 = ["x" + str(i) for i in range(n0)]

for n in layer0: 
    G.add_edge(n, "y") 
    G.node[n]["value"] = 0


nx.draw(G, node_size=1000, node_color='lightblue', with_labels=True)
```

```python
G = nx.DiGraph()

n0, n1 = 4, 3
layer0 = ["x" + str(i) for i in range(n0)]
layer1 = ["y" + str(i) for i in range(n1)]

for x in layer0: 
    for y in layer1: 
        G.add_edge(x, y) 
    G.node[x]["value"] = 0

nx.draw_shell(G, node_size=1000, node_color='lightblue', with_labels=True)
```

```python
G = nx.DiGraph()

n0, n1 = 4, 3
layer0 = ["x" + str(i) for i in range(n0)]
pos0 = {layer0[i]:(1, n0-i) for i in range(n0)}

layer1 = ["y" + str(i) for i in range(n1)]
pos1 = {layer1[i]:(3, n1-i) for i in range(n1)}

pos = pos0; pos.update(pos1)

for x in layer0: 
    for y in layer1: 
        G.add_edge(x, y) 
    G.node[x]["value"] = 0

nx.draw(G, pos = pos, node_size=1000, node_color='lightblue', with_labels=True)
```

```python
def getPos(n0,n1,kx = 1):
    if n0 > n1: ky0, ky1 = n0+1, n0+1-((n0-n1)/2)
    else: ky0, ky1 = n1+1-((n1-n0)/2), n1+1
    pos0 = {layer0[i]:(kx, ky0-i) for i in range(n0)}
    pos1 = {layer1[i]:(kx+1, ky1-i) for i in range(n1)}
    pos = pos0; pos.update(pos1)
    return pos

G = nx.DiGraph()

n0, n1 = 2, 5
layer0 = ["x" + str(i) for i in range(n0)]
layer1 = ["y" + str(i) for i in range(n1)]

pos = getPos(n0,n1)

for x in layer0: 
    for y in layer1: 
        G.add_edge(x, y) 
    G.node[x]["value"] = 0

nx.draw(G, pos = pos, node_size=1000, node_color='lightblue', with_labels=True)

```

```python
def addLayer(G, n0,n1,kx = 1):
    def getPos(n0,n1,kx = 1):
        if n0 > n1: ky0, ky1 = n0+1, n0+1-((n0-n1)/2)
        else: ky0, ky1 = n1+1-((n1-n0)/2), n1+1
        pos0 = {layer0[i]:(kx, ky0-i) for i in range(n0)}
        pos1 = {layer1[i]:(kx+1, ky1-i) for i in range(n1)}
        pos = pos0; pos.update(pos1)
        return pos
    layer0 = ["x" + str(i) for i in range(n0)]
    layer1 = ["y" + str(i) for i in range(n1)]
    
    for x in layer0: 
        for y in layer1: 
            G.add_edge(x, y) 
        G.node[x]["value"] = 0
    return getPos(n0,n1)

G = nx.DiGraph()
n0, n1 = 3, 1
pos = addLayer(G, n0,n1)
nx.draw(G, pos = pos, node_size=1000, node_color='lightblue', with_labels=True)
```
